```python
def convertToDT(s):
    if not s:
        return s
    dt = datetime.strptime(s,'%Y-%m-%d %H:%M:%S.%f') 
    return dt
```

```python
#ASSIGNMENT_START_TIME ASSIGNMENT_END_TIME PLACED_TIME DELIVERED_TIME
```

# Size 

**1. Size of issue (DE rejections) = 8048
**2. In 2739 cases, the assignment was given after the shift-time ended for the DE. 
**3. In other cases, DEs were willing to work overtime or they might have already used up their rejection limit of the day.
**4. In 5369 out of 8048 cases, the rejection is caused due to mismatch of respective zones of DE and the customer.

# Impact

**1. From the 2nd scatter plot, it can be observed that Type-1 rejections were made by the DEs who took high delivery time, shifting the Goldilocks value to upper side and thereby decreasing Customer engagements and rendering a bad customer experience.
**2. One of the major impacts is decrement in the customer segmentation prediction value and thereby resulting in the loss of revenue of Swiggy.
**3. The prepared food/meal will become stale due to the delay in assignment caused by several DEs rejecting it and re-heating food will render a bad customer experience like a too low Goldilocks value.
**4. The Entire assembly chain mechanism of many restaurants can get disturbed with DE rejecting it and resulting in the bad bridging gap between the Restaurants and Customers.
**5. Overall decrement in engagements with customers of particular zones/area, resulting in the loss of huge business revenue model.

# Cause

**1.Considering zone wise DE rejections, those maybe due to the mismatching in the zones of DE location and customer location causing hassle, frustration on the behalf of an experienced DE.
**2 Customers rejecting order on the basis of color, caste, religion and discriminating them therefore DE not interested in going to that particular zone.
**3 The assignment being given after the shift-time ended for the DE.

Yes, a machine learning algorithm can be applied for the mapping of DE with a desired assignment considering factors like Customer zone, DE Location, Shift-end time as important features.
